#### Are there columns having inappropriate data types?

```python
main_df[numerical_cols].dtypes
```

"YearsCode" and "YearsCodePro" of numerical_cols still have the object data types, so we will return the set of data types actually present in those columns.

```python
def open_object_dtype(s):
    
    dtypes = set(s.apply(type).unique())
    
    return dtypes
```

```python
print(open_object_dtype(main_df['YearsCode']))
print(open_object_dtype(main_df['YearsCodePro']))
```

Convert dtype of columns "YearsCode" and "YearsCodePro" to float.

```python
main_df['YearsCode'] = pd.to_numeric(main_df['YearsCode'], errors='coerce')
main_df['YearsCodePro'] = pd.to_numeric(main_df['YearsCodePro'], errors='coerce')
```

```python
main_df[numerical_cols].dtypes
```

#### With each numerical column, how are values distributed?
- What is the percentage of missing values?
- Min? max? Are they abnormal?

```python
def missing_ratio(s):
    return s.isna().mean() * 100
```

```python
nume_col_profiles_df = main_df[numerical_cols]
nume_col_profiles_df = nume_col_profiles_df.agg([missing_ratio, pd.Series.min, pd.Series.max])
nume_col_profiles_df.round(2)
```

Không có gì bất thường.

#### With each categorical column, how are values distributed?
- What is the percentage of missing values?
- How many different values? Show a few
- Are they abnormal?

```python
def diff_vals(s):
    return s.dropna().unique()
```

```python
cate_col_profiles_df = main_df[categorical_cols]
cate_col_profiles_df = cate_col_profiles_df.agg([missing_ratio, pd.Series.nunique, diff_vals])
cate_col_profiles_df = cate_col_profiles_df.rename(index={'nunique': 'num_diff_vals'})
cate_col_profiles_df
```

Không có điều gì bất thường.
